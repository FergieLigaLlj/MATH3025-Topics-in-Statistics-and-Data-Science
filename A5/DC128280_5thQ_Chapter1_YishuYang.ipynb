```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(n, num_samples):
  """Generates data according to the given pseudocode.

  Args:
    n: Degrees of freedom for the chi-square distribution.
    num_samples: Number of samples to generate.

  Returns:
    A numpy array of generated samples.
  """

  data = []
  for _ in range(num_samples):
    y = np.random.chisquare(df=n-1)
    z = np.random.normal(0, np.sqrt(y/(n-1)))
    data.append(z)
  return np.array(data)

# Parameters
n = 5  # Degrees of freedom
num_samples = 1000

# Generate data
data = generate_data(n, num_samples)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.7)
plt.title("Histogram of Generated Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
