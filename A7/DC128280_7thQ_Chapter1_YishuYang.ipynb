Here we choose g = U([-1,1]) so that g(Y) = 0.5 for any Y from [-1,1]. And we choose M = 5.
1. Sample a random quantity Y from the distribution g.
2. Sample a random quantity U from U([0, 1]).\
    if U <= f(Y)/(2.5): accept, ie. X = Y.\
    if U > f(Y)/(2.5): reject and repeat the procedure.\
end
It is worth mentioned that 2.5 here is the multiplication of M = 5 and g(Y) = 0.5.

```python
import random
import matplotlib.pyplot as plt
import numpy as np

def accept_reject(M):
    while True:
        x = 2 * random.random() - 1
        fx = 1 / (1 + abs(x - 2)**3)
        gx = 0.5
        a = fx / (M * gx)
        u = random.random()
        if u <= a:
            return x

# Simulate 1000 samples
samples = [accept_reject(5) for _ in range(1000)]

# Count the number of samples generated
sample_count = len(samples)

# Plot the histogram
plt.hist(samples, bins=30, density=True)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Histogram of Samples from the Target Distribution')
plt.show()

print("Number of samples generated:", sample_count)
```
