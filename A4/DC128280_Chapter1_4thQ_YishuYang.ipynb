```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(num_samples, p, mu1, sigma1, mu2, sigma2):
    """Generates data from a Gaussian Mixture Model.

    Args:
        num_samples: Number of samples to generate.
        p: Probability of sampling from the first Gaussian component.
        mu1: Mean of the first Gaussian component.
        sigma1: Standard deviation of the first Gaussian component.
        mu2: Mean of the second Gaussian component.
        sigma2: Standard deviation of the second Gaussian component.

    Returns:
        A numpy array of generated samples.
    """

    data = []
    for _ in range(num_samples):
        theta = np.random.binomial(1, p)
        if theta == 1:
            z = np.random.randn()
            x = mu1 + sigma1 * z
        else:
            z = np.random.randn()
            x = mu2 + sigma2 * z
        data.append(x)
    return np.array(data)

# Parameters
num_samples = 1000
p = 0.3  # Probability of sampling from the first Gaussian
mu1 = 0
sigma1 = 1
mu2 = 3
sigma2 = 2

# Generate data
data = generate_data(num_samples, p, mu1, sigma1, mu2, sigma2)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.7)
plt.title("Histogram of Generated Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
